- Анализ

```python
# свободные показатели
# кол-во выживших
titanic_data["survived"].sum()
```

```python
# средний возраст пассажиров
titanic_data["age"].mean()
```

```python
# минимальный возраст среди пассажиров
titanic_data["age"].min()
```

```python
# значение по столбцу "survived" при условии "age" == 0.42
# комбинация маскирования по строкам и обычной индексации
titanic_data.loc[titanic_data["age"] == 0.42, "survived"]
```

```python
# значения всех столбцов при условии возраст == 0,42
titanic_data.loc[titanic_data["age"] == 0.42, :]
```

```python
# значения некоторых столбцов при условии возраст == 80
# комбинация маскирования и прихотливой индексации
titanic_data.loc[titanic_data["age"] == 80, ["survived", "pclass", "sex", "fare", "alone"]]
```

```python
# сводные показатели по столбцу "age" при условии "survived" == 1
titanic_data.loc[titanic_data["survived"] == 1, "age"].mean()
```

```python
titanic_data.loc[titanic_data["survived"] == 0, "age"].mean()
```

```python
titanic_data.loc[titanic_data["survived"] == 0, "age"].min()
```

- GroupBy

```python
# "сгруппировать по"
# операция "разбиения", "применение", "объединение"
titanic_data.groupby("survived")
```

```python
titanic_data.groupby("survived").mean()
```

```python
titanic_data.groupby("survived")[["pclass", "age"]].mean()
```

```python
titanic_data.head()
```

```python
#
titanic_data.groupby("pclass")[["age"]].mean()
```

```python
titanic_data.groupby("who")[["age", "pclass"]].mean()
```

```python
# группировака данных по категориям "who", "survived"
titanic_data.groupby(["who", "survived"])[["pclass", "age"]].mean()
```

```python
titanic_data.groupby(["who", "survived", "pclass"])[["age"]].count()
```

```python
# применение нескольких агреирующих функций
titanic_data.aggregate(["mean", min, np.max])
```

```python
titanic_data.groupby(["who", "survived", "pclass"])[["age"]].aggregate(["mean", min, np.max, "count"])
```
