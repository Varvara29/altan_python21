```python
import numpy as np
import matplotlib.pyplot as plt
```

- простые линейные графики

```python
# одномерный массив (вектор) из 10 значений
x = np.arange(10)
x
```

```python
y = x
```

```python
#визуализация
plt.plot(x, y)
plt.show()
```

```python
# одномерный массив из 1000 значений
# линейная последовательность в диапазоне от 0 до 10
x = np.linspace(0, 10, 1000)
x
```

```python
# данные по оси x
x = np.linspace(-10, 10, 1000)

# определение функции
def sigmoid(arg):
    return 1/(1 + np.exp(-arg))

# вычисление значений по оси y
y = sigmoid(x)

#визуализация
plt.plot(x, y)
plt.show()
```

```python
# гиперболический тангенс
y = np.tanh(x)

plt.plot(x,y)
plt.show()
```

```python
# совмещение нескольких графиков в одном окне
x = np.linspace(-10, 10, 1000)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# цвет линии графиков, 7-8 способов

plt.plot(x, y1, color="red") # по названию
plt.plot(x, y2, color="k") #по первой букве цвета из формата rgb cmyk
color="ffa703" (linewidth=5) #по 16-ричному коду формата RGB (RGBA)
color=(0.0, 1.0, 0.5) #по формату RGB в виде кортежа со значениями в диапазоне от 0 до1
plt.show()

# 1-й способ

```

```python
# стиль линий

plt.plot(x, np.sin(x), linestyle='-') #"solid") #сплошная
plt.plot(x, np.sin(x - 1), linestyle='--') #="dashed") #штриховая
plt.plot(x, np.sin(x - 2), linestyle=':') #="") #пунктирная
plt.plot(x, np.sin(x - 3), linestyle='-.') #="dashed") #штрих-пунктирная
plt.show()
```

```python
#объединение настроек цвета и стиля линий

plt.plot(x, np.sin(x), "--g")
plt.plot(x, np.sin(x-1), ":r")

plt.show()
```

```python
# пределы осей
plt.plot(x, np.sin(x), "--g")

#индивидуальные пределы
#plt.xlim(-15, 15)
#plt.ylim(-1.5, 1.5)

plt.axis([-20, 20, -1.5, 1.5])

plt.show()
```

```python
#легенда графиков, подписи осей

plt.plot(x, np.sin(x), "--g", label="синус")
plt.plot(x, np.cos(x), ":r", label="косинус")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")

plt.axis([-20, 20, -2, 2])

plt.show()
```

```python
#заголовок графика

plt.plot(x, np.sin(x), "--g", label="синус")
plt.plot(x, np.cos(x), ":r", label="косинус")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")
plt.title("ГРАФИК")

plt.axis([-20, 20, -2, 2])

plt.show()
```
